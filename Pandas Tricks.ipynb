```python
fasla = pd.DataFrame(
    [[1, '1234', 'factory'],
     [2, '5678', 'Warehouse']],
    columns=['ID', 'Code', 'Name']
)
```

```python
fasla.head()
```

```python
fasla.melt(id_vars='ID',var_name='loc__Type',value_name='distance')
```

```python
fasla_melted = fasla.melt(id_vars='ID', var_name='loc__Type', value_name='distance')
sns.barplot(data=fasla_melted, x='ID', y='distance', hue='loc__Type')
```
